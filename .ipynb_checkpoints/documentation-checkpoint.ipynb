<h1>
    Documentation:
</h1>

<h4>
    <ul>
        <li>
            <h4> 
            We started with the idea of simply <em>simulating</em> data from a cloud server, instead of using a direct web API, since the inclusion and process of exposing the API itself is very complex, so to run the model on simulated data is a lot 
            easier. 
            </h4>
        </li> 
        <li>
            <h4> 
           The Z score is way to determine outliers/anomalies in a given dataset, plotting the z score derived from the simulated 
            dataset clearly shows spikes in data, this can be utilised to train a machine learning model to actually predict future anomalies. 
            </h4>
        </li> 
        <li>
            <h4> 
           At first, we decided to test on a linear regression model. Although the model worked, it wasn't suitable for the task 
                of classifying clustered data as seen from a working server data stream.
            </h4>
        </li> 
        <li>
            <h4> 
           In order to fix this and to further improve prediction accuracy, we decided to switch to a logistic regression model, 
            this further boosted prediction capabilities, since a logistic regression model is best designed to handle clustered scatterpoints. 
            </h4>
        </li>  
        <li>
            <h4> 
          The next phase included the process of integrating the AWS web server API with jupyter notebooks to directly run the python script containing the logistic regression algorithm. 
            </h4>
        </li> 
        <li>
            <h4> 
          The ML model gets trained on the data coming from the webserver, it uses real time data to create data-frames, clean the data set and train the model on it to predict anomalies.
            </h4>
        </li> 
        <li>
            <h4> 
         The model initially hallucinated, the spikes from the servers were much below the calculated z-scores and the z-score threshold was too high, causing the model to identify 0 anomalies, therby resulting in overfitting, regularisation and a clean dataset was needed. 
            </h4>
        </li> 
        <li>
            <h4> 
         We fixed this by appropriately lowering the z-score threshold, after which the model was able to identify about 12-13 anomalies from the current state of the server and based on the telemetry it provided after a purposeful 5 minute stress test.
            </h4> 
        </li>
        <li>
            <h4> 
         Finally, the model was able to predict anomalies with about 93% accuracy, and phase 3 was complete, with our first prototype.
            </h4>
        </li>





</h4>
